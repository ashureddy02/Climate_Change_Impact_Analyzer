```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/engineered_features.csv")
print("Rows:", len(df))
df.head()
```

```python
# Average rainfall per year
rainfall_yearly = df.groupby("Year")["Annual_Rainfall"].mean()

plt.figure()
rainfall_yearly.plot()
plt.xlabel("Year")
plt.ylabel("Avg Annual Rainfall (mm)")
plt.title("Annual Rainfall Trend in India")
plt.show()
```

```python
# Average temperature per year
temp_yearly = df.groupby("Year")["Avg_Temperature"].mean()

plt.figure()
temp_yearly.plot()
plt.xlabel("Year")
plt.ylabel("Avg Temperature (C)")
plt.title("Avg Temperature Trend in India")
plt.show()
```

```python
# Crop production graph trend over years
production_yearly = df.groupby("Year")["Total_Production"].mean()

plt.figure()
production_yearly.plot()
plt.xlabel("Year")
plt.ylabel("Avg Crop Production")
plt.title("Avg Crop Production Trend Over Years")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Annual_Rainfall"], df["Total_Production"])
plt.xlabel("Annual Rainfall (mm)")
plt.ylabel("Total Crop Production")
plt.title("Rainfall vs Crop Production")
plt.show()
```

```python
# Count temperature levels
temp_counts = df["Temp_Level"].value_counts()

plt.figure()
temp_counts.plot(kind="bar")
plt.xlabel("Temperature Level")
plt.ylabel("Count")
plt.title("Distribution of Temperature Levels")
plt.show()
```
